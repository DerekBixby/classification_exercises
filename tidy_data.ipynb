a. Read the cake_recipes table. This data set contains cake tastiness scores for combinations of different recipes, oven rack positions, and oven temperatures.

```python
url = "/Users/derekbixby/codeup-data-science/classification_exercises/untidy-data/cake_recipes.csv"

df_cake = pd.read_csv(url)
df_cake.head()
```

b. Tidy the data as necessary.

```python
df_cake [['recipe', 'position']] = df_cake['recipe:position'].str.split(':', expand=True)
df_cake
```

```python
df_cake = df_cake.drop(columns='recipe:position')
df_cake
```

c. Which recipe, on average, is the best? Recipe B

```python
new_cols = ['recipe', 'position', '225', '250', '275', '300']
df_cake=df_cake[new_cols]
df_cake
```

```python
df_cake_melt = df_cake.melt(id_vars=df_cake.columns[:2],
              var_name ='temperature',              
              value_name='rating')
df_cake_melt.head()
```

```python
df_cake_melt.groupby('recipe').rating.mean().max()
```
